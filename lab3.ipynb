```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./image/img.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
plt.show()
```

```python
def loc_thong_thap_ly_tuong(i, d):
    F = np.fft.fft2(i)
    M, N = i.shape
    H = np.zeros((M, N))
    for u in range(M):
        for v in range(N):
            t = ((u-M/2)**2+(v-N/2)**2)**0.5
            H[u, v] = (t <= d) if 1 else 0
    G = H*F
    plt.imshow(np.log(np.abs(np.fft.fftshift(np.fft.ifft2(G)))**2))
    plt.show()
```

```python
loc_thong_thap_ly_tuong(img_gray, 200)
```

```python
def loc_thong_thap_butterworth(i, d, n):
    F = np.fft.fft2(i)
    M, N = i.shape
    H = np.zeros((M, N))
    for u in range(M):
        for v in range(N):
            t = ((u-M/2)**2+(v-N/2)**2)**0.5
            H[u, v] = (1/(1+(t/d)**(2*n)))
    G = H*F
    plt.imshow(np.log(np.abs(np.fft.fftshift(np.fft.ifft2(G)))**2))
    plt.show()
```

```python
loc_thong_thap_butterworth(img_gray, 200, 2)
```

```python
def loc_thong_thap_gaussian(i, d):
    F = np.fft.fft2(i)
    M, N = i.shape
    H = np.zeros((M, N))
    for u in range(M):
        for v in range(N):
            t = ((u-M/2)**2+(v-N/2)**2)**0.5
            H[u, v] = np.exp(-t**2/(2*(d**2)))
    G = H*F
    plt.imshow(np.log(np.abs(np.fft.fftshift(np.fft.ifft2(G)))**2))
    plt.show()
```

```python
loc_thong_thap_gaussian(img_gray, 200)
```
